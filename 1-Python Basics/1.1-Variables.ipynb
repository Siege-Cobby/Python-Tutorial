## Variables

Variables are fundamental elements in programming used to store data that can be referenced and manipulated in a program. In Python, variables are created when you assign a value to them, and they do not need explicit declaration to reserve memory space. The declaration happens automatically when you assign a value to a variable.

```python
## Declaring & Assigning Variables
age = 34
height = 1.75
name = "Siege"
is_student = True

## Printing Variables
print("Age:", age)
print("Height:", height)
print("Name:", name)
```

```python
## Naming Conventions
## should be descriptive and follow conventions (e.g., snake_case for variables)
student_name = "Siege"
student_age = 34

## must start with a letter or underscore, cannot start with a number
# 1st_name = "Siege"  # Invalid variable name
# .name = "Siege"  # Invalid variable name
first_name = "Siege"  # Valid variable name

## variable names are case-sensitive
First_Name = "Siege"  # Different variable from first_name
print("First Name:", first_name)
print("First Name (case-sensitive):", First_Name)
```

```python
## Understanding Variable Types
# Python has several built-in data types, including:
# - int: Integer numbers (e.g., 1, 2, 3)
# - float: Decimal numbers (e.g., 1.5, 2.7)
# - str: Strings (e.g., "Hello", "World")
# - bool: Boolean values (True or False)
# You can check the type of a variable using the type() function
age = 34 # Integer
height = 1.75 # Float
name = "Siege" # String
is_student = True # Boolean

print("Type of age:", type(age))
print("Type of height:", type(height))
print("Type of name:", type(name))
print("Type of is_student:", type(is_student))
```

```python
## Type Conversion
# You can convert between different data types using built-in functions:
# - int(): Converts to an integer
# - float(): Converts to a float
# - str(): Converts to a string
# - bool(): Converts to a boolean
age=34
print(type(age))  # Output: <class 'int'>
print(age_str)
age_str = str(age)  # Convert integer to string
print(type(age_str))  # Output: <class 'str'>
```

```python
## Dynamic Typing
# Python is a dynamically typed language, which means you can change the type of a variable by assigning a new value of a different type to it.
variable = 10  # variable is an integer
print("Variable:", variable, "Type:", type(variable))
variable = "Now I'm a string"  # variable is now a string
print("Variable:", variable, "Type:", type(variable))
```

```python
## Input and Output
# You can take user input using the input() function, which always returns a string. You can convert it to the desired type if needed.
user_name = input("Enter your name: ")  # Takes user input as a string
print("Hello, " + user_name + "!")  # Output: Hello, [user_name]!
```

```python
### Simple Calculator Example
# This example demonstrates how to take user input, perform type conversion, and output results.
num1 = float(input("Enter the first number: "))  # Takes user input as a string and converts it to a float
num2 = float(input("Enter the second number: "))  # Takes user input as a string and converts it to a float

result = num1 + num2  # Adds the two numbers
difference = num1 - num2  # Subtracts the second number from the first
product = num1 * num2  # Multiplies the two numbers
quotient = num1 / num2  # Divides the first number by the second

print("The sum is:", result)
print("The difference is:", difference)
print("The product is:", product)
print("The quotient is:", quotient)
```
