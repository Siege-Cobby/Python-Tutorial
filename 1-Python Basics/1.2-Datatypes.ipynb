## Data Types

#### Definition
In Python, data types are categories that tell the computer what kind of value a variable holds and what you can do with it.

Think of them like different "modes" for your data:

- Integers (int): Whole numbers used for counting. (e.g., 10, -3)
- Floats (float): Numbers with decimal points. (e.g., 3.14, 10.0)
- Strings (str): Text or characters wrapped in quotes. (e.g., "Hello")
- Booleans (bool): Logical values that are either True or False.
They determine the types of operations that can be performed on the data, and the amount of memory needed to store the data.

#### Importance of Data Types in Programming
- Ensures data is stored in an efficient way.
- Operation Rules: Data types define what you can actually do. You can multiply two integers, but you can't multiply a string by a boolean.
- Error Prevention: They act as a safety rail. If you try to calculate the square root of a word like "Apple," the data type system flags this as an error before the program crashes.
- Data Integrity: They ensure your data stays in the correct format (e.g., ensuring a "Price" field stays as a number and doesn't accidentally become text).

```python
## Integers
# Integers are whole numbers without a decimal point.
age = 34
type(age)  # Output: <class 'int'>
print("Age:", age, "Type:", type(age))
```

```python
## Floats
# Floats are numbers that contain a decimal point.
height = 5.6
print(height)
print(type(height))  # Output: <class 'float'>
```

```python
## Strings
# Strings are sequences of characters enclosed in quotes.
name = "Siege"
print(name)
print(type(name))  # Output: <class 'str'>
```

```python
## Booleans
# Booleans represent truth values and can be either True or False.
is_student = True
print(is_student)
print(type(is_student))  # Output: <class 'bool'>
```

```python
## Booleans Continued
a=10  # Output: True
b=10
print(a==b)  # Output: True
c=5
print(a<c)  # Output: False
```
