```python
## Arithmetic Operations
a = 10
b = 5

addition = a + b  # Output: 15
subtraction = a - b  # Output: 5
multiplication = a * b  # Output: 50
division = a / b  # Output: 2.0
floor_division = a // b  # Output: 2
modulus = a % b  # Output: 0
exponentiation = a ** b  # Output: 100000

print("Addition:", addition)
print("Subtraction:", subtraction)
print("Multiplication:", multiplication)
print("Division:", division)
print("Floor Division:", floor_division)
print("Modulus:", modulus)
print("Exponentiation:", exponentiation)
```

```python
## Comparison Operators
x = 10
y = 20

x==y  # Output: False
x!=y  # Output: True
x>y  # Output: False
x<y  # Output: True
x>=y  # Output: False
x<=y  # Output: True

print("x == y:", x==y)
print("x != y:", x!=y)
print("x > y:", x>y)
print("x < y:", x<y)
print("x >= y:", x>=y)
print("x <= y:", x<=y)
```

```python
## Logical Operators And, Or, Not
a = True
b = False

and_result = a and b  # Output: False
or_result = a or b  # Output: True
not_a = not a  # Output: False
not_b = not b  # Output: True

print("a and b:", and_result) 
print("a or b:", or_result)
print("not a:", not_a)
print("not b:", not_b)
```
